```python
#TASK:2

# Count duplicate rows
duplicates = sales_df.duplicated().sum()

print("Number of duplicate rows found:", duplicates)

# Remove duplicates
sales_df = sales_df.drop_duplicates()

print("Number of duplicate rows removed:", duplicates)


# Fill missing values in quantity with 0
sales_df['quantity'] = sales_df['quantity'].fillna(0)

# Remove rows where amount is missing
sales_df = sales_df.dropna(subset=['amount'])

print("\nShape of cleaned DataFrame:", sales_df.shape)



# Convert sale_date to datetime format
sales_df['sale_date'] = pd.to_datetime(sales_df['sale_date'])

# Convert amount to float
sales_df['amount'] = sales_df['amount'].astype(float)

# Check data types
print("\nUpdated Data Types:")
print(sales_df.dtypes)
```

```python
#TASK:3


# Merge sales and products data on product_id
merged_df = pd.merge(sales_df, products_df, on='product_id', how='left')

# Merge the above result with stores data on store_id
final_df = pd.merge(merged_df, stores_df, on='store_id', how='left')

print("Final Merged DataFrame:")
print(final_df)


# Calculate total revenue
final_df['total_revenue'] = final_df['quantity'] * final_df['price']

print("\nMean Revenue:", np.mean(final_df['total_revenue']))
print("Maximum Revenue:", np.max(final_df['total_revenue']))
print("Minimum Revenue:", np.min(final_df['total_revenue']))


city_revenue = final_df.groupby('city')['total_revenue'].sum()

# Sort in descending order
city_revenue = city_revenue.sort_values(ascending=False)

print("\nTotal Revenue Generated Per City:")
print(city_revenue)
```

```python
#TASK:4

# Create connection with SQLite database
conn = sqlite3.connect('retailmart.db')

# Load DataFrame into SQL table
final_df.to_sql('retail_sales', conn, if_exists='replace', index=False)

print("Data loaded successfully into 'retail_sales' table.")

# Check the table
query = "SELECT * FROM retail_sales LIMIT 5"
result = pd.read_sql(query, conn)

print("\nFirst 5 rows from retail_sales table:")
print(result)
```

```python
# Find top 3 products
query = """
SELECT
    product_name,
    SUM(quantity) AS total_quantity_sold
FROM retail_sales
GROUP BY product_name
ORDER BY total_quantity_sold DESC
LIMIT 3;
"""

top_products = pd.read_sql(query, conn)

print("Top 3 Best-Selling Products:")
print(top_products)
```

```python
#TASK:5
# Find total revenue per store per day

query = """
SELECT
    store_name,
    sale_date,
    SUM(total_revenue) AS daily_revenue
FROM retail_sales
GROUP BY store_name, sale_date
ORDER BY sale_date, daily_revenue DESC;
"""

revenue_per_store = pd.read_sql(query, conn)

print("Total Revenue per Store per Day:")
print(revenue_per_store)
```

```python
#TASK:6

def run_pipeline():

    try:
        # Step 1: Load Data
        sales_df = pd.read_csv("sales_data.csv")
        products_df = pd.read_csv("products.csv")
        stores_df = pd.read_csv("stores.csv")

        print("All files loaded successfully.\n")

        # Step 2: Clean Data

        # Remove duplicate rows
        duplicates = sales_df.duplicated().sum()
        sales_df = sales_df.drop_duplicates()
        print(f"{duplicates} duplicate rows removed.")

        # Fill missing quantity values with 0
        sales_df['quantity'] = sales_df['quantity'].fillna(0)

        # Remove rows where amount is missing
        sales_df = sales_df.dropna(subset=['amount'])

        # Convert data types
        sales_df['sale_date'] = pd.to_datetime(sales_df['sale_date'])
        sales_df['amount'] = sales_df['amount'].astype(float)

        print("Data cleaning completed.\n")

        # Step 3: Transform Data

        # Merge DataFrames
        merged_df = pd.merge(
            sales_df,
            products_df,
            on='product_id',
            how='left'
        )

        final_df = pd.merge(
            merged_df,
            stores_df,
            on='store_id',
            how='left'
        )

        # Create total_revenue column
        final_df['total_revenue'] = (
            final_df['quantity'] * final_df['price']
        )

        print("Data transformation completed.\n")

        # Step 4: Load into SQLite Database

        conn = sqlite3.connect("retailmart.db")

        final_df.to_sql(
            "retail_sales",
            conn,
            if_exists='replace',
            index=False
        )

        print("Data loaded successfully into SQLite database.\n")

        print("Pipeline executed successfully!")

        return final_df

    except FileNotFoundError as e:
        print("Error: One or more CSV files are missing.")
        print(e)

    except Exception as e:
        print("An unexpected error occurred:")
        print(e)

final_df = run_pipeline()
```
